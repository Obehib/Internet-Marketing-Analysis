- Create and visualize the symmetric and weighted networks of Bollywood actors.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx
```

# Symmetric Network of Bollywood Actors

```python
G_symmetric = nx.Graph()
G_symmetric.add_edge('Amitabh Bachchan','Abhishek Bachchan')
G_symmetric.add_edge('Amitabh Bachchan','Admir Khan')
G_symmetric.add_edge('Amitabh Bachchan','Akshay Kumar')
G_symmetric.add_edge('Amitabh Bachchan','Dev Anand')
G_symmetric.add_edge('Abhishek Bachchan','Admir Khan')
G_symmetric.add_edge('Abhishek Bachchan','Akshay Kumar')
G_symmetric.add_edge('Abhishek Bachchan','Dev Anand')
G_symmetric.add_edge('Dev Anand','Admir Khan')
```

```python
nx.draw_networkx(G_symmetric)
```

# Weighted Network of Bollywood Actors

```python
G_weighted = nx.Graph()
G_weighted.add_edge('Amitabh Bachchan','Abhishek Bachchan', weight=25)
G_weighted.add_edge('Amitabh Bachchan','Admir Khan', weight=8)
G_weighted.add_edge('Amitabh Bachchan','Akshay Kumar', weight=11)
G_weighted.add_edge('Amitabh Bachchan','Dev Anand', weight=1)
G_weighted.add_edge('Abhishek Bachchan','Admir Khan', weight=4)
G_weighted.add_edge('Abhishek Bachchan','Akshay Kumar',weight=7)
G_weighted.add_edge('Abhishek Bachchan','Dev Anand', weight=1)
G_weighted.add_edge('Dev Anand','Admir Khan',weight=1)
```

```python
plt.figure(figsize=(10,7))
nx.draw_networkx(G_weighted)
```

# Admir Khan's node degree

```python
nx.degree(G_symmetric, 'Admir Khan')
```

```python
nx.clustering(G_symmetric, 'Admir Khan')
```

```python
nx.clustering(G_weighted, 'Admir Khan')
```

# Admir Khan's Shortest Path to Akshay Kumar

```python
nx.shortest_path(G_symmetric, 'Admir Khan', 'Akshay Kumar')
```

# Length of the Path

```python
nx.shortest_path_length(G_symmetric, 'Admir Khan', 'Akshay Kumar')
```

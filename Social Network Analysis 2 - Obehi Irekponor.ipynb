Visualize the network data
- Show the top five node labels with highest Between Centrality. 
- Show the top five node labels with other two centrality measures:  Eigenvector Centrality and Degree Centrality

```python
import networkx as nx
```

```python
G_fb = nx.read_edgelist("facebook_combined.txt", create_using = nx.Graph(), nodetype=int)
```

```python
print(nx.info(G_fb))
```

```python
pos = nx.spring_layout(G_fb)
betCent = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
node_color = [20000.0 * G_fb.degree(v) for v in G_fb]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off')
```

# Top Five Node Labels with Highest Between Centrality.

```python
sorted(betCent, key=betCent.get, reverse=True)[:5]
```

# Top Five Node Labels with Highest Eigenvector Centrality

```python
cent_eig = nx.eigenvector_centrality(G_fb)
```

```python
sorted(cent_eig, key=cent_eig.get, reverse=True)[:5]
```

# Top Five Node Labels with Highest Degree Centrality

```python
cent_degree = nx.degree_centrality(G_fb)
```

```python
sorted(cent_degree, key=cent_degree.get, reverse=True)[:5]
```
